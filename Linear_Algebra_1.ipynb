```python
import numpy as np

A = np.array([
    [1,  2,  4],
    [3, -1,  5],
    [-4, -3, -11]
])

rank = np.linalg.matrix_rank(A)
print("Rank:", rank)
```

# Find the Rank
**How to interpret:**

*   If rank(A) == 3 → span is all of R³
*   If rank(A) == 2 → span is a plane in R³
*   If rank(A) == 1 → span is a line

For your vectors, you’ll get:
rank(A) = 3 (and det(A) = 1), so the span is R³.


```python
import numpy as np

A = np.array([
    [2, 1, 1],
    [1, 0, 0],
    [1, 2, 1]
], dtype=float)

print("rank(A) =", np.linalg.matrix_rank(A))
```

```python
import sympy as sp

v1 = sp.Matrix([2, 1, 1])
v2 = sp.Matrix([1, 0, 2])
v3 = sp.Matrix([1, 0, 1])

A = sp.Matrix.hstack(v1, v2, v3)  # put vectors as columns

print("A =")
sp.pprint(A)

print("\nrank(A) =", A.rank())
print("det(A)  =", A.det())

rref_A, pivots = A.rref()
print("\nRREF(A) =")
sp.pprint(rref_A)
print("Pivot columns =", pivots)
```

# Step 1: Define the matrix

```python
import numpy as np

A = np.array([
    [2, 1],
    [-3, 6]
])
```

# Step 2: Compute eigenvalues

```python
eigenvalues = np.linalg.eigvals(A)
print("Eigenvalues:", eigenvalues)
```

# Step 3: Build the characteristic polynomial manually
For a 2×2 matrix:
Step 3: Build the characteristic polynomial manually

For a 2×2 matrix:
p(λ) = λ² − (trace)λ + det⁡(A)


```python
trace = np.trace(A)
det = np.linalg.det(A)

print("Characteristic polynomial:")
print(f"λ² - {trace}λ + {det}")
```

# Using Sympy

```python
import sympy as sp

A = sp.Matrix([
    [2, 1, 0],
    [1, 3, 1],
    [0, 1, 2]
])

λ = sp.symbols('λ')
char_poly = A.charpoly(λ)

print(char_poly.as_expr())
```
